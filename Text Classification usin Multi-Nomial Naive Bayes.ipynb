os.path.join(path, *paths)
Join one or more path components intelligently. The return value is the concatenation of path and any members of *paths with exactly one directory separator (os.sep) following each non-empty part except the last, meaning that the result will only end in a separator if the last part is empty. If a component is an absolute path, all previous components are thrown away and joining continues from the absolute path component.
X = [] # an element of X is represented as (filename,text)
Y = [] # an element of Y represents the newsgroup category of the corresponding X element

```python
data = {}
X = []
Y = []
for folder in os.listdir('20_newsgroups'):
    for file in os.listdir(os.path.join('20_newsgroups',folder)):
        with open(os.path.join('20_newsgroups',folder,file) , "r") as f:
            X.append((file , f.read()))
            Y.append(folder)
```

```python
#Creating set of stopwords

from nltk.corpus import stopwords
from string import punctuation

sw = list(stopwords.words('english'))
sw += list(punctuation)
sw += ['subject:','from:', 'date:', 'newsgroups:', 'message-id:', 'lines:', 'path:', 'organization:', 
            'would', 'writes:', 'references:', 'article', 'sender:', 'nntp-posting-host:', 'people', 
            'university', 'think', 'xref:', 'cantaloupe.srv.cs.cmu.edu', 'could', 'distribution:', 'first', 
            'anyone','world', 'really', 'since', 'right', 'believe', 'still', 'keywords','expires','approved'
            ,'archive','name','to:','or:','telephone:','fax:','reply-to:',"that's",'followup-to:',"we're",
            "let's", "what's","max>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'","going","using",]
```

```python
#Removing stop words from words list

def filter_words(words):
    useful_words = [w for w in words if w not in sw]
    return useful_words
```

```python
#converting sentences into tokens
from nltk.tokenize import word_tokenize

vocab = {} 
for i in range(len(X)):
    words = word_tokenize(X[i][1].lower())
    useful_words = filter_words(words)
    for j in useful_words:
        if j in vocab:
            vocab[j]+=1
        else:
            vocab[j] = 1
```

```python
vocab['atheism']
```

```python
from operator import itemgetter
vocab = dict(sorted(vocab.items() , key = itemgetter(1) , reverse = True))
```

```python
print(max(vocab.values())+1)
```

```python
import matplotlib.pyplot as plt

#y-axis
size = max(vocab.values())+1
No_of_words = [0] * size
#x-axis
Word_frequency = [i for i in range(size)]
#create list of count of no of words with specific frequency
for i in vocab:
    No_of_words[vocab[i]] +=1

#plotting the graph    
plt.plot(Word_frequency,No_of_words)
plt.axis([1, 25, 0, 20000])
plt.xlabel("Frequency")
plt.ylabel("No of words")
plt.grid()
plt.show()
```

```python
#Feature Words
feature = []
cutoff_freq = 80
for key in vocab:
    if vocab[key] >=cutoff_freq:
        feature.append(key)
len(feature)
```

X_train Dataset and Y_Train Dataset for naive bayes implementation
We will be building a counting vector

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.25, random_state=0)
```

1.) Created an array of zeros of shape no_of_feature * total_X_Train elements
2.) And then for each element each word of it is checked whether it matches any feature
3.) if yes then that particural cell for that word of that element is incremented

```python
X_train_dataset = np.zeros((len(X_train),len(feature)))

for i in range(len(X_train)):
    word_list = word_tokenize(X_train[i][1].lower())
    for word in word_list:
        if(word in feature):
            X_train_dataset[i][feature.index(word)] += 1
```

```python
print(X_train_dataset)
```

```python
X_test_dataset = np.zeros((len(X_test),len(feature)))

for i in range(len(X_test)):
    word_list = word_tokenize(X_test[i][1].lower())
    for word in word_list:
        if(word in feature):
            X_test_dataset[i][feature.index(word)] += 1
```

```python
len(Y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
clf = MultinomialNB()
clf.fit(X_train_dataset,Y_train)
Y_test_pred = clf.predict(X_test_dataset)
sklearn_score_train = clf.score(X_train_dataset,Y_train)
print("Sklearn's score on training data :",sklearn_score_train)
sklearn_score_test = clf.score(X_test_dataset,Y_test)
print("Sklearn's score on testing data :",sklearn_score_test)
print("Classification report for testing data :-")
print(classification_report(Y_test, Y_test_pred))
```
